```python
#Question 1
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data/crypto.csv', parse_dates=[1])
bitcoin = df[df['Asset_Name'] == 'Bitcoin']
ethereum = df[df['Asset_Name'] == 'Ethereum']
plt.plot(bitcoin['timestamp'], bitcoin['Close'])
plt.plot(ethereum['timestamp'], ethereum['Close'])
plt.show()
```

```python
#All graphs on one plot
asset_list=df["Asset_Name"].unique().tolist()
for asset in asset_list:
    asset_df = df[df['Asset_Name'] == asset]
    plt.plot(asset_df['timestamp'], asset_df['Close'], label = asset)
plt.legend()
```

```python
#Logarithmic form of all graphs
asset_list=df["Asset_Name"].unique().tolist()
for asset in asset_list:
    asset_df = df[df['Asset_Name'] == asset]
    plt.plot(asset_df['timestamp'], asset_df['Close'], label = asset)
    plt.yscale('log')
plt.legend()
```
