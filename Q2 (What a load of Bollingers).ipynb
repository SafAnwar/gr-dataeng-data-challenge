```python
#Question 2ai
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data/crypto.csv', parse_dates=[1])
bitcoin = df[df['Asset_Name'] == 'Bitcoin'].copy()
bitcoin["Rolling"] = bitcoin['Close'].rolling(3).mean()
plt.plot(bitcoin['timestamp'], bitcoin['Rolling'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
period = int(input("Enter a number to indicate the number of days in average: "))
df = pd.read_csv('data/crypto.csv', parse_dates=[1])
bitcoin = df[df['Asset_Name'] == 'Bitcoin']
plt.plot(bitcoin['timestamp'], bitcoin['Close'])
bitcoin = df[df['Asset_Name'] == 'Bitcoin'].copy()
bitcoin["Rolling"] = bitcoin['Close'].rolling(period).mean()
plt.plot(bitcoin['timestamp'], bitcoin['Rolling'])
plt.show()
```
